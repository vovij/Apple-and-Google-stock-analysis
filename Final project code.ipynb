```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import csv
from datetime import datetime
import numpy as np

# Read the file and extract the data
dates1 = []
dates2 = []
y1 = []
y2 = []

with open('AAPL.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        date = datetime.strptime(row[0], '%Y-%m-%d')
        dates1.append(date)
        y1.append(float(row[1]))
        
with open('google-stock-dataset-Monthly.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        date = datetime.strptime(row[1], '%Y-%m-%d')
        dates2.append(date)
        y2.append(float(row[2]))


plt.plot(dates1, y1, label = "Apple stock price", color = 'red')
plt.plot(dates2, y2, label = "Google stock price", color = 'blue')
plt.xlabel("Time, in years")
plt.ylabel("Price of individual stock, in dollars")

plt.title('Apple and Google stock prices, as on the first day of each month during the preiod of May 2013 to April 2023')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import csv
from datetime import datetime
import numpy as np

# Read the file and extract the data
datesv1 = []
datesv2= []
yv1 = []
yv2= []

with open('AAPL.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        date = datetime.strptime(row[0], '%Y-%m-%d')
        datesv1.append(date)
        yv1.append(float(row[6])) #/30.343 is the average number of day in a month
        
with open('google-stock-dataset-Monthly.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        date = datetime.strptime(row[1], '%Y-%m-%d')
        datesv2.append(date)
        yv2.append(float(row[6])) #/30.343 is the average number of day in a month


plt.plot(datesv1, yv1, label = "Apple trading volume", color = 'red')
plt.plot(datesv2, yv2, label = "Google trading volume", color = 'blue')
plt.xlabel("Time, in years")
plt.ylabel("Number of stocks traded in a month, in billions")

plt.title('Monthly trading volume of Apple and Google stocks')
plt.legend()
```

```python

x=np.array(y1)
y=np.array(y2)
m = sum(np.multiply(x-np.mean(x),y-np.mean(y)))/ sum(np.multiply(x-np.mean(x),x-np.mean(x)))
#print(m)
q = np.mean(y)-m*np.mean(x)
#print(m,q)


plt.plot(y1,y2, 'o', color = 'blue')

# Labelling 

plt.xlabel("Apple stock price")
plt.ylabel("Google stock price")
plt.title("Scatter plot")
plt.axis([0,200,0,180])

lin_reg = m*x+q
plt.plot(y1, lin_reg, color = 'orange')

print("For linear regression with Apple stock price on x axis (as shown on plot), m={:.4f} and q={:.2f}".format(m,q))
x=np.array(y2)
y=np.array(y1)
m = sum(np.multiply(x-np.mean(x),y-np.mean(y)))/ sum(np.multiply(x-np.mean(x),x-np.mean(x)))
q = np.mean(y)-m*np.mean(x)
print("For linear regression with Google stock price on x axis, m={:.4f} and q={:.1f}".format(m,q))


# Display
plt.title('Scatter plot of Apple and Google stock prices')
plt.show()

```

```python
import numpy as np
x = y1
y = y2
cov = sum(np.multiply(x-np.mean(x),y-np.mean(y)))/(len(x)-1)
coeff = cov / (np.std(x,ddof=1)*np.std(y,ddof=1))
print ('Correlation coefficient =', coeff)
```

```python
import numpy as np
from scipy.stats import pearsonr

x = y1
y = y2

correlation_coefficient, p_value = pearsonr(x, y)

print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")
```
